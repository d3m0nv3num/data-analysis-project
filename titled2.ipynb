```python
myfreqtable = data['Please rate the course content'].value_counts()
myfreqtable2 = data['Do you feel these training sessions helped you in building confidence to speak up ?'].value_counts()
myfreqtable3 = data['Please rate the study material provided to follow up with the classes'].value_counts()
mykeys = myfreqtable.keys()
mykeys2 = myfreqtable2.keys()
mykeys3 = myfreqtable3.keys()
myvals = myfreqtable.values
myvals2 = myfreqtable2.values
myvals3 = myfreqtable3.values
myfreqtableor = pd.DataFrame({' ratings': mykeys, 'content quality': myvals,'confidence build':myvals})
myfreqtableor2 = pd.DataFrame({' ratings': mykeys3, 'follow up material':myvals3 })


```

```python
w=0.3
bar1= np.arange(len(mykeys))
bar2=[i+w for i in bar1]
bar3=[i+w for i in bar2]
mykeys6 = [i+w for i in mykeys]
mykeys7 = [i-w for i in mykeys2]
plt.bar(mykeys3,myvals3,.3,label="Follow up material",color='red')
plt.bar(mykeys6,myvals,w,label="Content quality ",color='maroon')
plt.bar(mykeys7,myvals2,.3,label="Confidence build",color='gold')
plt.xlabel("Ratings(Out of five)")
plt.ylabel("Number of students")
plt.title("Rating of course")
plt.legend()
plt.show()
```

2.As clearly visible from graph content quality varies from good to a 5/5(with a significantly high 5/5 ratings than 4 or 3) rating hence good content quality,also follow up material is of good quality although confidence build can be improved upon
